```python
# • Calculate total marks
df["Total"]=df['English']+df['Maths']+df['Science']
print(df)
```

```python
# Calculate percentage
total_marks=300 # 3*100
df['Percentage']=(df["Total"]/total_marks)*100
print(df)
```

```python
# Assign grades based on percentage
def grade(percentage):
    if percentage >= 90:
        return "A+"
    elif percentage >= 80:
        return "A"
    elif percentage >= 70:
        return "B"
    elif percentage >= 60:
        return "C"
    elif percentage >= 50:
        return "D"
    else:
        return "F"

df["Grade"] = df["Percentage"].apply(grade)
df
```

```python
# • Top performing student

top_std=df.loc[df['Total'].idxmax()]
top_std
```

```python
# • Lowest performing student
lowest_std=df.loc[df['Total'].idxmin()]
lowest_std
```

```python
# • Average marks in each subject
sub=['English','Maths','Science']
av=df[sub].mean()
print(av)
```

```python
# • Highest attendance
high_att=df.loc[df['Attendance'].idxmax()]
high_att
```

```python
#  Students scoring below average
average=df['Total'].mean()
below_avg=df[df["Total"]<average]
below_avg
```

```python
# • Gender-wise average performance
avg_per=df.groupby('Gender')['Total'].mean()
avg_per
```
